## Trying to fill NaN values of IMDB_score, IMDB_votes, TMDB_popularity, TMDB_score

```python
tdf=pd.merge(rel,con_gen, on=['id','Type'],how='inner')
tdf=pd.merge(tdf,crd,on='id',how='inner')
```

```python
t=tdf.dropna()
xyz=t.groupby(['name','role','Country Name','Release Year']).mean()
xyz=xyz.reset_index()
xyz['person_id']=xyz['person_id'].astype('int')
xyz['Release Year']=xyz['Release Year'].astype('int')
```

```python
# xyz.to_excel('C:\\Users\\adity\\OneDrive\\Documents\\Python Scripts\\Netflix\\Actors_Score.xlsx')
```

```python
dfn=t.groupby(['name']).mean()
dfn=dfn.reset_index()
dfn['person_id']=dfn['person_id'].astype('int')
dfn['Release Year']=dfn['Release Year'].astype('int')
```

```python
tlt_e=tlt.copy()
```

### Filling IMDB score

```python
tempc=crd[['person_id','id']]
tempc=pd.merge(tempc,dfn, on='person_id',how='left')
for i in tlt_e.index:
    if (tlt_e['imdb_score'][i]>=0)==False:
        t=tempc[tempc['id']==tlt_e['id'][i]]
        t=t[['id','person_id','IMDB_score']]
        t=t.dropna()
        xyz=t['person_id'].unique()
        abc=0
        for j in xyz:
            t=tempc[tempc['person_id']==j]
            t=t[['id','person_id','IMDB_score']]
            if sum(t['IMDB_score'])>=0:
                abc=abc+(sum(t['IMDB_score'])/len(t))
        if len(xyz)!=0:
            abc=abc/len(xyz)
        
        if abc!=0:
            tlt_e.at[i,'imdb_score']=abc
```

### Filling IMDB votes

```python
for i in tlt_e.index:
    if (tlt_e['imdb_votes'][i]>=0)==False:
        t=tempc[tempc['id']==tlt_e['id'][i]]
        t=t[['id','person_id','IMDB_score']]
        t=t.dropna()
        xyz=t['person_id'].unique()
        abc=0
        for j in xyz:
            t=tempc[tempc['person_id']==j]
            t=t[['id','person_id','IMDB_votes']]
            if sum(t['IMDB_votes'])>=0:
                abc=abc+(sum(t['IMDB_votes'])/len(t))
        if len(xyz)!=0:
            abc=abc/len(xyz)
        
        if abc!=0:
            tlt_e.at[i,'imdb_votes']=abc
```

### Filling TMDB Popularity

```python
for i in tlt_e.index:
    if (tlt_e['tmdb_popularity'][i]>=0)==False:
        t=tempc[tempc['id']==tlt_e['id'][i]]
        t=t[['id','person_id','TMBD_populatiry']]
        t=t.dropna()
        xyz=t['person_id'].unique()
        abc=0
        for j in xyz:
            t=tempc[tempc['person_id']==j]
            t=t[['id','person_id','TMBD_populatiry']]
            if sum(t['TMBD_populatiry'])>=0:
                abc=abc+(sum(t['TMBD_populatiry'])/len(t))
        if len(xyz)!=0:
            abc=abc/len(xyz)
        
        if abc!=0:
            tlt_e.at[i,'tmdb_popularity']=abc
```

### Filling TMDB score

```python
for i in tlt_e.index:
    if (tlt_e['tmdb_score'][i]>=0)==False:
        t=tempc[tempc['id']==tlt_e['id'][i]]
        t=t[['id','person_id','TMBD_score']]
        t=t.dropna()
        xyz=t['person_id'].unique()
        abc=0
        for j in xyz:
            t=tempc[tempc['person_id']==j]
            t=t[['id','person_id','TMBD_score']]
            if sum(t['TMBD_score'])>=0:
                abc=abc+(sum(t['TMBD_score'])/len(t))
        if len(xyz)!=0:
            abc=abc/len(xyz)
        
        if abc!=0:
            tlt_e.at[i,'tmdb_score']=abc
```
